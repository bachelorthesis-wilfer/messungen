```python
import csv
import json
import matplotlib.pyplot as plt
from datetime import datetime
```

# Spring Blocking


```python
# path params
json_path = './sbj-ilo.json'
csv_file1 = "./output/10min-spring-blocking-jar.csv"
plot_image_file = "./output/10min-spring-blocking-jar.png"

with open(json_path, 'r') as file1:
    loaded_data = json.load(file1)

# Extract Time and Average
data1 = loaded_data["PowerDetail"]
time_values1 = [datetime.fromisoformat(entry["Time"].replace("Z", "+00:00")) for entry in data1]
average_values1 = [entry["Average"] for entry in data1]


with open(csv_file1, mode='w', newline='') as file1:
    writer1 = csv.writer(file1)
    writer1.writerow(["Time", "Average"])  # Write header row
    writer1.writerows(zip(time_values1, average_values1))  # Write data rows

print(f'Data from the JSON file has been written to "{csv_file1}".')

# Create two subplots sharing the same X-axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the PowerDetail data on the left Y-axis
ax1.plot(time_values1, average_values1, marker='.', linestyle='-', label="Power Average", color='blue')
ax1.set_xlabel('Time')
ax1.set_ylabel('Power Average', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(140, 260)

# Title and legend
plt.title('10min Spring Blocking Platform Jar')
plt.legend(loc='upper right')

# Format x-axis labels as dates
plt.xticks(rotation=45)

plt.savefig(plot_image_file)

# Show the plot
plt.tight_layout()
plt.show()

print(f'Combined plot has been saved as "{plot_image_file}".')
```

```python
# From GPT zusammengewurschtelt

import json
import csv
import matplotlib.pyplot as plt
from datetime import datetime

# Define the paths and colors for each dataset
datasets = [
    {"json_path": './sbj-ilo.json', "csv_file": "./output/10min-sbj.csv", "color": "blue", "label": "SBJ"},
    {"json_path": './srg-ilo.json', "csv_file": "./output/10min-srg.csv", "color": "red", "label": "SRG"},
    {"json_path": './srj-ilo.json', "csv_file": "./output/10min-srj.csv", "color": "green", "label": "SRJ"},
    {"json_path": './sbvj-ilo.json', "csv_file": "./output/10min-sbvj.csv", "color": "purple", "label": "SBVJ"},
    {"json_path": './sbg-ilo.json', "csv_file": "./output/10min-sbg.csv", "color": "orange", "label": "SBG"}
]

# Create figure and axis for the combined plot
fig, ax1 = plt.subplots(figsize=(10, 5))

for dataset in datasets:
    json_path = dataset["json_path"]
    csv_file = dataset["csv_file"]
    color = dataset["color"]
    label = dataset["label"]

    # Load JSON data
    with open(json_path, 'r') as file1:
        loaded_data = json.load(file1)

    # Extract Time and Average
    data1 = loaded_data["PowerDetail"]
    time_values1 = [datetime.fromisoformat(entry["Time"].replace("Z", "+00:00")) for entry in data1]
    average_values1 = [entry["Average"] for entry in data1]

    # Write the data to the CSV file
    with open(csv_file, mode='w', newline='') as file1:
        writer1 = csv.writer(file1)
        writer1.writerow(["Time", "Average"])  # Write header row
        writer1.writerows(zip(time_values1, average_values1))  # Write data rows

    print(f'Data from the JSON file has been written to "{csv_file}".')

    # Plot the data
    ax1.plot(time_values1, average_values1, marker='.', linestyle='-', label=label, color=color)

# Customize the plot
ax1.set_xlabel('Time')
ax1.set_ylabel('Power Average')
ax1.tick_params(axis='y')
ax1.set_ylim(140, 260)

# Title and legend
plt.title('10min Combined Spring')
plt.legend(loc='upper right')

# Format x-axis labels as dates
plt.xticks(rotation=45)

# Save the combined plot
plot_image_file = "./output/combined_plot.png"
plt.savefig(plot_image_file)

# Show the plot
plt.tight_layout()
plt.show()

print(f'Combined plot has been saved as "{plot_image_file}".')
```

# Mean und Average


```python
import pandas as pd
csv_files = [
    "./output/10min-sbj.csv",
    "./output/10min-srg.csv",
    "./output/10min-srj.csv",
    "./output/10min-sbvj.csv",
    "./output/10min-sbg.csv"
]


# Schleife ber die CSV-Dateien
for csv_file in csv_files:
    # CSV-Datei laden
    df = pd.read_csv(csv_file)
    
    # Berechnungen
    mean_value = df["Average"].mean()
    median_value = df["Average"].median()
    num_records = df["Average"].count()
    
    # Ausgabe
    print(f"Datei: {csv_file}")
    print(f"Anzahl der Datenstze: {num_records}")
    print(f"Mittelwert (Mean) von 'Average': {mean_value}")
    print(f"Median von 'Average': {median_value}")
    print("----")
```
