```python
import csv
import json
import matplotlib.pyplot as plt
from datetime import datetime
```

# Quarkus Blocking

```python

# path params
json_path = './qbj-ilo.json'
csv_file1 = "./output/10min-quarkus-blocking-jar.csv"
plot_image_file = "./output/10min-quarkus-blocking-jar.png"

with open(json_path, 'r') as file1:
    loaded_data = json.load(file1)

# Extract Time and Average
data1 = loaded_data["PowerDetail"]
time_values1 = [datetime.fromisoformat(entry["Time"].replace("Z", "+00:00")) for entry in data1]
average_values1 = [entry["Average"] for entry in data1]


with open(csv_file1, mode='w', newline='') as file1:
    writer1 = csv.writer(file1)
    writer1.writerow(["Time", "Average"])  # Write header row
    writer1.writerows(zip(time_values1, average_values1))  # Write data rows

print(f'Data from the JSON file has been written to "{csv_file1}".')

# Create two subplots sharing the same X-axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the PowerDetail data on the left Y-axis
ax1.plot(time_values1, average_values1, marker='.', linestyle='-', label="Power Average", color='blue')
ax1.set_xlabel('Time')
ax1.set_ylabel('Power Average', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(140, 260)

# Title and legend
plt.title('Quarkus Blocking JAR 10 Min')
plt.legend(loc='upper right')

# Format x-axis labels as dates
plt.xticks(rotation=45)

plt.savefig(plot_image_file)

# Show the plot
plt.tight_layout()
plt.show()

print(f'Combined plot has been saved as "{plot_image_file}".')
```

```python

import json
import csv
import matplotlib.pyplot as plt
from datetime import datetime

# Define the paths and colors for each dataset
datasets = [
    {"json_path": './qbj-ilo.json', "csv_file": "./output/10min-qbj.csv", "color": "blue", "label": "QBJ"},
    {"json_path": './qrg-ilo.json', "csv_file": "./output/10min-qrg.csv", "color": "red", "label": "QRG"},
    {"json_path": './qrj-ilo.json', "csv_file": "./output/10min-qrj.csv", "color": "green", "label": "QRJ"},
    {"json_path": './qbvj-ilo.json', "csv_file": "./output/10min-qbvj.csv", "color": "purple", "label": "QBVJ"},
    {"json_path": './qbg-ilo.json', "csv_file": "./output/10min-qbg.csv", "color": "orange", "label": "QBG"}
]

# Create figure and axis for the combined plot
fig, ax1 = plt.subplots(figsize=(10, 5))

for dataset in datasets:
    json_path = dataset["json_path"]
    csv_file = dataset["csv_file"]
    color = dataset["color"]
    label = dataset["label"]

    # Load JSON data
    with open(json_path, 'r') as file1:
        loaded_data = json.load(file1)

    # Extract Time and Average
    data1 = loaded_data["PowerDetail"]
    time_values1 = [datetime.fromisoformat(entry["Time"].replace("Z", "+00:00")) for entry in data1]
    average_values1 = [entry["Average"] for entry in data1]

    # Write the data to the CSV file
    with open(csv_file, mode='w', newline='') as file1:
        writer1 = csv.writer(file1)
        writer1.writerow(["Time", "Average"])  # Write header row
        writer1.writerows(zip(time_values1, average_values1))  # Write data rows

    print(f'Data from the JSON file has been written to "{csv_file}".')

    # Plot the data
    ax1.plot(time_values1, average_values1, marker='.', linestyle='-', label=label, color=color)

# Customize the plot
ax1.set_xlabel('Time')
ax1.set_ylabel('Power Average')
ax1.tick_params(axis='y')
ax1.set_ylim(140, 260)

# Title and legend
plt.title('10min Combined Quarkus')
plt.legend(loc='upper right')

# Format x-axis labels as dates
plt.xticks(rotation=45)

# Save the combined plot
plot_image_file = "./output/combined_plot.png"
plt.savefig(plot_image_file)

# Show the plot
plt.tight_layout()
plt.show()

print(f'Combined plot has been saved as "{plot_image_file}".')
```
